# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load environment variables
%load_ext dotenv
%dotenv
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, PowerTransformer
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.naive_bayes import GaussianNB 

import os
import sys

# Standard libraries
import pandas as pd
import numpy as np
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
# Define 'X' variable. Assuming all 'area' is the response variable and all other variables are predictor variables
X = fires_dt.drop(columns=['area'])
X
```

```python
y = fires_dt['area']
```

```python
#Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, RobustScaler
from sklearn.preprocessing import FunctionTransformer
```

```python
numeric_cols = ['ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
categorical_cols = ['month', 'day']
```

```python
preproc1 = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(), categorical_cols)])

preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
preproc2 = ColumnTransformer(
   transformers=[
      ('num', Pipeline([
        ('scaler', StandardScaler()),
        ('log', FunctionTransformer(np.log1p, validate=True))  # Apply log transformation
       ]), numeric_cols),
       ('cat', OneHotEncoder(), categorical_cols)
    ])

preproc2
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
baseline_regressor = Ridge()
advanced_regressor = RandomForestRegressor(random_state=42)
```

```python
# Pipeline A = preproc1 + baseline
pipeline_a = Pipeline([
    ('preprocessing', ColumnTransformer([
        ('numeric', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),  # Impute missing values with median
            ('scaler', StandardScaler())  # StandardScaler for scaling numeric features
        ]), numeric_cols),
        ('categorical', Pipeline([
            ('onehot', OneHotEncoder(handle_unknown='ignore'))  # OneHot encoding for categorical variables
        ]), categorical_cols)
    ])),
    ('regressor', Ridge())  # Regressor (Ridge model)
])
pipeline_a
```

```python
# Pipeline B = preproc2 + baseline
pipeline_b = Pipeline([
    ('preprocessing', ColumnTransformer([
        ('numeric', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),  # Impute missing values with median
            ('scaler', RobustScaler())  # Using RobustScaler instead of StandardScaler
        ]), numeric_cols),
        ('categorical', Pipeline([
            ('onehot', OneHotEncoder(handle_unknown='ignore'))  # OneHot encoding for categorical variables
        ]), categorical_cols)
    ])),
    ('regressor', Ridge())  # Regressor (Ridge model)
])
pipeline_b
```

```python
# Pipeline C = preproc1 + advanced model
pipeline_c = Pipeline([
    ('preprocessing', preproc1),
    ('regressor', advanced_regressor)
])
pipeline_c
```

```python
# Pipeline D = preproc2 + advanced model
pipeline_d = Pipeline([
    ('preprocessing', preproc2),
    ('regressor', advanced_regressor)
])
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
```

```python
param_grid_a = {'regressor__alpha': [0.1, 1, 10, 100]
}
grid_search_a = GridSearchCV(pipeline_a,param_grid_a, cv=5)
```

```python
param_grid_b = {'regressor__alpha': [0.1, 1, 10, 100]
}
grid_search_b = GridSearchCV(pipeline_b,param_grid_b, cv=5)
```

```python
param_grid_c = {'regressor__n_estimators': [50, 100, 200, 300],
    'regressor__max_depth': [10, 20, 30, None]
}
grid_search_c = GridSearchCV(pipeline_c,param_grid_c, cv=5)
```

```python
param_grid_d = {
    'regressor__n_estimators': [50, 100, 200, 300],
    'regressor__max_depth': [10, 20, 30, None]
}
grid_search_d = GridSearchCV(pipeline_d,param_grid_d, cv=5)
```

# Evaluate

+ Which model has the best performance?

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# fitting models
grid_search_a.fit(X_train, y_train)
grid_search_b.fit(X_train, y_train)
grid_search_c.fit(X_train, y_train)
grid_search_d.fit(X_train, y_train)

```

```python
# Predictions for each model
y_pred_a = grid_search_a.best_estimator_.predict(X_test)
y_pred_b = grid_search_b.best_estimator_.predict(X_test)
y_pred_c = grid_search_c.best_estimator_.predict(X_test)
y_pred_d = grid_search_d.best_estimator_.predict(X_test)
```

```python
# model evaluation using RMSE and MAE
rmse_a = np.sqrt(mean_squared_error(y_test, y_pred_a))
mae_a = mean_absolute_error(y_test, y_pred_a)
rmse_a, mae_a
```

```python
rmse_b = np.sqrt(mean_squared_error(y_test, y_pred_b))
mae_b = mean_absolute_error(y_test, y_pred_b)
rmse_b, mae_b
```

```python
rmse_c = np.sqrt(mean_squared_error(y_test, y_pred_c))
mae_c = mean_absolute_error(y_test, y_pred_c)
rmse_c, mae_c
```

```python
rmse_d = np.sqrt(mean_squared_error(y_test, y_pred_d))
mae_d = mean_absolute_error(y_test, y_pred_d)
rmse_d, mae_d
```

If we are evaluating on the basis of rmse: Model d will be best performing as it has the lowest rmse

SHAP values provide a way to explain how each feature contributes to a model's prediction. For the best-performing model, we can select an observation from the test set and use SHAP values to break down the prediction. SHAP values show how each feature shifts the prediction from the baseline value to the final prediction. For example, in a forest fire prediction model, features like temperature, humidity, and wind speed could have significant impacts on predicting the area affected by the fire. The SHAP analysis for a specific test observation would show which features contributed most to the predicted area, highlighting whether higher wind speeds or lower humidity led to a larger fire prediction. Across the entire training set, the most important features would be those that consistently have a high impact, while the least important features would be those with minimal impact.

We can calculate SHAP values and evaluate which of the features has the lowest SHAP value. This lowest SHAP value features would be the ones with minimal impact. We can then use feature elimination and assess the model's performance using methods such as cross validation. If removing affects the models performance we can see that it is a critical and important feature.
